```
import pandas as pd
import matplotlib.pyplot as plt
```

## Baixando os dados


```
! wget https://www.gstatic.com/covid19/mobility/Region_Mobility_Report_CSVs.zip
! unzip  Region_Mobility_Report_CSVs.zip -d data
```

```
df = pd.read_csv('/content/data/2020_BR_Region_Mobility_Report.csv')
```

```
df.info()
```

```
df.describe()
```

```
df['sub_region_1'].unique()
```

```
plt.style.use('default')
df.groupby(['date'])['residential_percent_change_from_baseline'].mean().round().plot(label='Brasil', linestyle='dashed')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
plt.xlabel('Dias')
plt.ylabel('Porcentagem de variação')
plt.title('Tendência em área residencial')
```

Dando inicio a análise, vamos iniciar pelo estado de São Paulo

```
df_sao_paulo = df.loc[df['sub_region_1'].isin(['State of São Paulo'])]
df_sao_paulo
```

```
plt.style.use('default')
df_sao_paulo.groupby(['date'])['residential_percent_change_from_baseline'].mean().round().plot(label='São Paulo', linestyle='dashed')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
plt.xlabel('Dias')
plt.title('Porcentagem de variação')
```
